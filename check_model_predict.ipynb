#### Load Yolo model predict image to verify output label.

```python
from ultralytics import YOLO
from pathlib import Path
model_weight = "weights/yolo/best_yolo11m_pencilbag.pt"
images_path = Path("datasets/robot-detection/data/images/pencilbag_task/")

valid_suffixes = {".png", ".jpg", ".jpeg", ".bmp"}
images = [f for f in images_path.iterdir() if f.is_file() and f.suffix.lower() in valid_suffixes]


label_map = {
  "0": "pencilbag",
  "1": "pig-zipper",
  "2": "simple-zipper",
  "3": "pen"
}
```

```python
model_predict = YOLO(model_weight)



results2 = model_predict.predict(
    source=images_path, 
    save=True, 
    project=str(Path.cwd() / "inference_results"), 
    name="pencilbag",
    conf=0.5
)
```
