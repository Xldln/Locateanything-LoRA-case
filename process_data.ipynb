```python
import os
from pathlib import Path
from tqdm import tqdm

dataset_path = Path("datasets/robot-detection/data/images")
task_list_dir = [d for d in dataset_path.iterdir() if d.is_dir()]
valid_suffixes = {".png", ".jpg", ".jpeg", ".bmp"}
for task_dir in task_list_dir:
    task_name = task_dir.name.split("_")[0]

    images = [f for f in task_dir.iterdir() if f.is_file() and f.suffix.lower() in valid_suffixes]
    
    count = 1

    for old_path in tqdm(images, desc=f"processing {task_dir.name}"):
        new_name = f"{task_name}_{count}.png"
        new_path = old_path.with_name(new_name)
        
        old_path.rename(new_path)
        count += 1


print("process end.")
```
